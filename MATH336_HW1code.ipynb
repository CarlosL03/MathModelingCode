## 3.7(b)

```python
from scipy.stats import linregress

def plot_trendline(start_year, end_year, color, label):
    yr = np.arange(start_year, end_year + 1)  # create a sequence of years
    subset = jan_tmin_data[(jan_tmin_data['YEAR'] >= start_year) & (jan_tmin_data['YEAR'] <= end_year)]
    le = subset['TMIN (F)'].astype(float) # get interval of temperature
    slope, intercept, r_value, p_value, std_err = linregress(yr, le)
    plt.plot(yr, slope*yr + intercept, color=color, linestyle='--', linewidth=2, label=label)
    return (slope)
```

```python
plt.figure(figsize=(12,8))
plt.plot(jan_tmin_data["YEAR"], jan_tmin_data["TMIN (F)"], color='blue', label='January Tmin')
trends = []
a = plot_trendline(1951, 2010, 'red', 'Trend 1951-2010')
trends.append(a)
b = plot_trendline(1961, 2010, 'green', 'Trend 1961-2010')
trends.append(b)
c = plot_trendline(1971, 2010, 'orange', 'Trend 1971-2010')
trends.append(c)
d = plot_trendline(1981, 2010, 'purple', 'Trend 1981-2010')
trends.append(d)

plt.xlabel('Year')
plt.ylabel('TMIN (F)')
plt.title("Cuyamaca January Tmin Time Series wi")
plt.legend()
plt.show()
```

```python
def calculateYearMeans(df):
    means = []
    summ = 0
    for i in range(0, 1644):
        summ += df.iloc[i]
        if (i+1) % 12 == 0:
            means.append(summ/12.0)
            summ = 0
    
    return means       
```
