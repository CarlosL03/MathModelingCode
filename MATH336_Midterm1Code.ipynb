```python
#find dataframe row containing life expectancy data of France
FranceData = df[df['Country Name'] == "France"]
#find columns for life-exp data from 1971 to 2010
FranceIntervData = FranceData.iloc[0, 15:55]
print(FranceIntervData)
```

```python
FranceLifeArr = np.array(FranceIntervData)
a, b = np.polyfit(yrs, FranceLifeArr.astype(float), 1)
print(f"Change rate of life expectancy for France from 1971-2010 = {a*10} years per decade")
#Change rate of life expectancy for France from 1971-2010 = 2.489019814671683 years per decade
```

## 4.ii

```python
# Read data file as csv from class data folder file path
file_path2 = '/Users/carlos_lopez/MATH336/data/EarthTemperatureData.csv'
tempdf = \
    pd.read_csv(file_path2,
               header=0, index_col=0)
#print first 5 rows
print(tempdf.head(5))
#print column name
print(tempdf.keys())
```

```python
#extract global January temp column data as an array (y-axis)
JanTemp = np.array(tempdf.loc['1850':'2015', 'JAN'])
print(JanTemp)
```

```python
#create yearly range array fit to January temp range (x-axis)
years = np.arange(1850,2016)

plt.figure(figsize=(10,6))
#4.ii(b) change curve to blue and use linewidth=3
plt.plot(years, JanTemp.astype(float), color = 'blue', linewidth=3)

plt.xlabel('Year', size = 14)
plt.ylabel('Temperature [deg C]', size = 14)
#4.ii(a) correct title with name and January
plt.title("Carlos Lopez's plot of January temperature anomalies", 
          size = 16, fontweight="bold")

#4.ii(c) compute linear trend of January anomalies
temTrend = np.polyfit(years, JanTemp.astype(float), 1)
print("Linear trend =", round(temTrend[0], 6), "deg C/year")
#Linear trend = 0.005018 deg C/year
print("Linear trend =", round(temTrend[0] * 100, 6), "deg C/century")
#Linear trend = 0.501759 deg C/century

#4.ii(d) plot linear trend: change color to blue and use linewidth=3
plt.plot(years, temTrend[1] + temTrend[0]*years, color='blue', linewidth=3)

#4.ii(e)
plt.text(1850, 0.55, "January trend = 0.5018 deg/century", size = 18)

plt.grid()
plt.tight_layout()
plt.show()
```

```python
#4.ii(g) plot the histogram of the January temperature anomalies
plt.figure(figsize=(10,6))

plt.hist(JanTemp, edgecolor='k', color = 'crimson', bins= 31)
plt.xlabel('Temperature anomalies [deg C].', size = 14)
plt.ylabel('Frequency', size = 14)
plt.title("Histogram of Global January Temperature Anomalies", 
          size = 16, fontweight="bold")
plt.tick_params(direction='in',labelsize=12)
plt.show()
```
